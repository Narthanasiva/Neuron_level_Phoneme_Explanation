# **Import libraries and define paths**

```python
import pandas as pd
import matplotlib.pyplot as plt

# üìÅ Load your saved CSV files from Drive
f1_df = pd.read_csv('/content/drive/MyDrive/00_RESEARCH_MSC_00/Final_Phonetic_Identification/03_Evaluation_metrics_of_probes/f1_scores.csv')        # update the path
acc_df = pd.read_csv('/content/drive/MyDrive/00_RESEARCH_MSC_00/Final_Phonetic_Identification/03_Evaluation_metrics_of_probes/accuracy_scores.csv') # update the path

# üîÅ Define features to loop over
features = ["voiced", "fricative", "nasal"]
colors = ["blue", "green", "red"]
```

# **F1 score VS Layers**

```python
# -----------------------------------------
# üìä 1. Plot F1 Score vs Layer
# -----------------------------------------
plt.figure(figsize=(10, 6))
for feat, color in zip(features, colors):
    plt.plot(f1_df["layer"], f1_df[feat], label=feat.capitalize(), color=color, marker='o')

plt.title("F1 Score vs WavLM Layer")
plt.xlabel("WavLM Layer")
plt.ylabel("F1 Score")
plt.xticks(range(13))
plt.ylim([0, 1.0])  # üëà Set y-axis limits here
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

# **Accuracy VS Layers**

```python
# -----------------------------------------
# üìä 2. Plot Accuracy vs Layer
# -----------------------------------------
plt.figure(figsize=(10, 6))
for feat, color in zip(features, colors):
    plt.plot(acc_df["layer"], acc_df[feat], label=feat.capitalize(), color=color, marker='o')

plt.title("Accuracy vs WavLM Layer")
plt.xlabel("WavLM Layer")
plt.ylabel("Accuracy")
plt.xticks(range(13))
plt.ylim([0, 1.0])  # üëà Set y-axis limits here
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
